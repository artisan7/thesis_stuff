```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('covid-chestxray-dataset/metadata.csv')
```

```python
df
```

```python
df.info()
```

```python
filt = (df['modality'] == 'X-ray') & (df['view'].isin(['AP', 'PA']))
df = df.loc[filt]
```

```python
df['finding'].unique()
```

```python
pneumonia_findings = {
    'Pneumonia',
    'Pneumonia/Viral/SARS',
    'Pneumonia/Fungal/Pneumocystis',
    'Pneumonia/Bacterial/Streptococcus',
    'Pneumonia/Bacterial/Chlamydophila',
    'Pneumonia/Bacterial/Klebsiella',
    'Pneumonia/Bacterial/Legionella',
    'Pneumonia/Lipoid', 'Pneumonia/Viral/Varicella',
    'Pneumonia/Bacterial',
    'Pneumonia/Bacterial/Mycoplasma',
    'Pneumonia/Viral/Influenza',
    'Tuberculosis',
    'Pneumonia/Viral/Influenza/H1N1',
    'Pneumonia/Viral/Herpes ',
    'Pneumonia/Fungal/Aspergillosis',
    'Pneumonia/Aspiration',
    'Pneumonia/Bacterial/Nocardia',
    'Pneumonia/Viral/MERS-CoV',
    'Pneumonia/Bacterial/Staphylococcus/MRSA'
}

df['finding'].replace(pneumonia_findings, 'Pneumonia', inplace=True)
df['finding'].replace('Pneumonia/Viral/COVID-19', 'COVID-19', inplace=True)
df['finding'].replace('todo', np.nan, inplace=True)

df['finding'].unique()
```

```python
df.dropna(how='all', subset=['finding'], inplace=True)
df
```

```python
output_df = df[['patientid', 'finding', 'view', 'modality', 'filename']]
output_df
```

```python
output_df.to_csv('clean_data.csv')
```
